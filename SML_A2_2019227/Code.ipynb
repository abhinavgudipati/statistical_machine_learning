```python
import numpy as np
from numpy import *

def perform_simplex_algo(A,b,C,m:int,n:int,mini=False,maxi=False):
	basic_vars = np.zeros((m)) # m slack variables for m constraints
	Ab = np.zeros((m,m))
	Cb = np.zeros((m));
	An = np.zeros((m,n)) # beacuse A = m X (m+n)
	Cn = np.zeros((n));
	entering_var_idx = -1
	exiting_var_idx = -1

	reduced_cost_vector = np.zeros((n+m)) # I will be filling only those positions which are currently in basic_vars, rest will be 0
	optimal_value = 0
	optimal_solution = 0	
	eps = 1e-10
	iteration_number = 0

#initialising Ab
	for i in range(m):
		for j in range(m):
			Ab[i,j] = A[i,n+j]

#initialising An
	for i in range(m):
		for j in range(n):
			An[i,j] = A[i,j]

#initialising basic_vars with slack variables
	for i in range(m):
		basic_vars[i] = n+i

#initialising Cn and Cb respectively
	for i in range(n):
		Cn[i] = C[i]
	# for i in range(m):
		# Cb[i] = C[i]
######################################################################################################################################################			
	# if(maximum_val_rc==0): #optimal value aagyi agar maxi=True ke lie maximum_val_rc 0 hi rahi kyuki Cn mei sab negative aagye
	# 	return 
######################################################################################################################################################
	it=0
	while(True):
		# print("Ab",Ab)
		it+=1;
		print()
		print("-------------------------------Iteration ",it," -------------------------------------------")
		print("Cb",Cb)
		reduced_cost_vector = C-np.dot(Cb.transpose(),np.dot(np.linalg.inv(Ab),A))
		print("reduced_cost_vector",reduced_cost_vector)
		if(maxi==True):
			maximum_val_rc = 0
			for i in range(0,n+m):
				# print(-eps)
				if((reduced_cost_vector[i])>eps and maximum_val_rc < reduced_cost_vector[i]):  # (reduced_cost_vector[i])>eps This conditionI have added warna fir float mei dikkat aarhi hai
					maximum_val_rc = reduced_cost_vector[i]
					entering_var_idx = i
			print("maximum_val_rc",maximum_val_rc)
			if(maximum_val_rc==0): #optimal value aagyi agar maxi=True ke lie maximum_val_rc 0 hi rahi kyuki Cn mei sab negative aagye
				break 
		elif(mini==True):
			minimum_val_rc = 0
			for i in range(0,n+m):
				if((reduced_cost_vector[i])<(-eps) and minimum_val_rc > reduced_cost_vector[i]):
					minimum_val_rc = reduced_cost_vector[i]
					entering_var_idx = i
			print("minimum_val_rc",minimum_val_rc)
			if(minimum_val_rc==0): #optimal value aagyi agar mini=True ke lie minimum_val_rc 0 hi rahi kyuki Cn mei sab positive aagye
				break 
		iteration_number += 1
		mini_bratio = 1e12
		exiting_var_idx = -1
		for i in range(m):
			if(np.dot(np.linalg.inv(Ab),A)[i, entering_var_idx] > 0):
				curr_bratio = np.dot(np.linalg.inv(Ab),b)[i]/np.dot(np.linalg.inv(Ab),A)[i, entering_var_idx]
				if(mini_bratio > curr_bratio):
					exiting_var_idx = i
					mini_bratio = curr_bratio
		if(exiting_var_idx==-1):
			print("Unbounded problem")
			return optimal_value, optimal_solution, reduced_cost_vector
		basic_vars[exiting_var_idx] = entering_var_idx
		# skipping the steo of sorting the basic_vars

		for i in range(m):
			for j in range(m+n):
				if(j==basic_vars[i]):
					Ab[:,i] = A[:,j]
					Cb[i] = C[j]
		optimal_solution = np.dot(np.linalg.inv(Ab),b)
		optimal_value = np.dot(Cb,optimal_solution)
	return optimal_value, optimal_solution, reduced_cost_vector

A = np.array([
	[1,2, 2, 1, 0, 0],
	[2,1, 2, 0, 1, 0],
	[2,2, 1, 0, 0, 1],
	])
b = np.array([[20],[20], [20]])
C = np.array([10,12,12,0,0,0])

optimal_value, optimal_solution, reduced_cost = perform_simplex_algo(A,b,C,3,3,False,True)
print("****************************************************************************************************************")
print("Final optimal value = ",optimal_value)
print("Final optimal solution = ")
print(optimal_solution)
# print(reduced_cost)
# A.length() -> number of constraints = m
# C.length() -> number of variables in Z functions = n 

```
